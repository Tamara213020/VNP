```python
from tsfresh import extract_features
from tsfresh.utilities.dataframe_functions import make_forecasting_frame
import pandas as pd
import yfinance as yf

# Пример со S&P 500 индексот
sp500 = yf.download('^GSPC', start='2020-01-01', end='2024-01-01')['Adj Close']

# Подготовка на податоците за tsfresh
df = pd.DataFrame({'timestamp': sp500.index, 'value': sp500.values})
df['id'] = 'SP500'  # Доделување на идентификатор за временската серија

# Екстракција на карактеристики со tsfresh
extracted_features = extract_features(df, column_id='id', column_sort='timestamp')

# Прикажување на првите неколку карактеристики
print(extracted_features.head())
```

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(x=df['timestamp'], y=df['value'], mode='lines', name='S&P 500'))
fig.update_layout(title='S&P 500 Index Over Time', xaxis_title='Date', yaxis_title='Index Value')
fig.show()
```

```python
from tsfresh import extract_features
from tsfresh.utilities.dataframe_functions import make_forecasting_frame
import pandas as pd
import yfinance as yf

selected_series = {
    'S&P 500': '^GSPC',
    'Bitcoin': 'BTC-USD',
    'WTI Crude Oil': 'CL=F',
    'Gold': 'GC=F',
    'Natural Gas': 'NG=F'
}

start_date = '2020-01-01'
end_date = '2024-01-01'

# Зачувување на податоците за секоја временска серија во dictionary
data = {}
for name, ticker in selected_series.items():
    series_data = yf.download(ticker, start=start_date, end=end_date)['Adj Close']
    df = pd.DataFrame({'timestamp': series_data.index, 'value': series_data.values})
    df['id'] = name  # Доделување на идентификатор за временската серија
    data[name] = df

extracted_features = {}
for name, df in data.items():
    features = extract_features(df, column_id='id', column_sort='timestamp')
    extracted_features[name] = features

    # Прикажување на првите неколку карактеристики за секоја серија
    print(f"Features for {name}:")
    print(features.head())
    print()



```

Оваа анализа со tsfresh ги извлекува различните статистички и временски карактеристики од секоја временска серија. Тие карактеристики можат да бидат корисни за разбирање на особините на временските серии, како што се варијабилноста, променливоста, енергијата, средната промена, спектралните ентропии итн.

Временските карактеристики што се извлекуваат со tsfresh нудат детален увид во особините на временските серии, што може да помогне во разбирањето на нивната структура, трендови, варијации и други аспекти.

Оваа анализа со tsfresh овозможува детално проучување и карактеризација на временските серии преку автоматско извлекување на различни статистички и временски карактеристики, што е од значење за напредна анализа и моделирање на временски серии.

```python
# Користење на plotly за визуелизација

import plotly.graph_objects as go

fig = go.Figure()

for name, df in data.items():
    fig.add_trace(go.Scatter(x=df['timestamp'], y=df['value'], mode='lines', name=name))

fig.update_layout(title='Time Series Comparison', xaxis_title='Date', yaxis_title='Value')
fig.show()
```

```python
import matplotlib.pyplot as plt

# Создавање на временски график за секоја временска серија со matplotlib
plt.figure(figsize=(14, 8))

for name, df in data.items():
    plt.plot(df['timestamp'], df['value'], label=name)

plt.title('Time Series Comparison')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Создавање на временски график за секоја временска серија со seaborn
plt.figure(figsize=(14, 8))

for name, df in data.items():
    sns.lineplot(x='timestamp', y='value', data=df, label=name)

plt.title('Time Series Comparison')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

Во нашиот случај кога се бираат пет различни временски серии и се проучуваат податоците со користење на tsfresh, изборот меѓу Plotly и Matplotlib зависи од следниве аспекти:

Интерактивност: Ако ви требаат високо интерактивни визуелизации каде што корисниците можат да зумираат, селектираат и добијат детали, тогаш Plotly е подобар избор поради своите напредни интерактивни можности.

Едноставност и брзина на креирање: За брзо создавање на стандардни временски графици без многу интерактивност, Matplotlib може да биде подобар избор поради што не е сложен и има широка употреба.

Поддршка за сложени анализи: Бидејќи ги анализираме временските серии користејќи tsfresh за екстракција на различни карактеристики, тогаш и двата графички пакети (Plotly и Matplotlib) се погодни. Ваквите анализи често може да вклучуваат потреба за преглед на различни аспекти од податоците во различни формати, каде што Plotly може да биде посебно корисен за подетална интеракција и анализа.

Споредувајќи ги овие аспекти, сепак, Matplotlib според мене е подобар избор за брзо создавање на стандардни графици и за едноставна анализа на податоци.
